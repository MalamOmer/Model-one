```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'your_dataset.csv' with the path to your dataset
data = pd.read_csv('dataset1.csv')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())

# Display basic information about the dataset
print("\nDataset Info:")
print(data.info())

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Example Analysis: Correlation Matrix
print("\nCorrelation Matrix:")
print(data.corr())

# Plotting example: Histogram of a column (replace 'column_name' with your column)
column_name = 'N'  # Change this to a valid column name
if column_name in data.columns:
    plt.figure(figsize=(8, 6))
    data[column_name].hist(bins=30, color='blue', alpha=0.7)
    plt.title(f"Histogram of {column_name}")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.show()
else:
    print(f"Column '{column_name}' not found in the dataset.")

# Example Analysis: Grouping and aggregation
if column_name in data.columns:
    print("\nGroup by Example:")
    group_analysis = data.groupby(column_name).mean()
    print(group_analysis)

# Scatter plot for two numerical columns (replace 'x_column' and 'y_column' with valid column names)
x_column = 'p'  # Change this to a valid column name
y_column = 'k'  # Change this to a valid column name
if x_column in data.columns and y_column in data.columns:
    plt.figure(figsize=(8, 6))
    plt.scatter(data[x_column], data[y_column], alpha=0.7, color='green')
    plt.title(f"Scatter Plot of {p} vs {k}")
    plt.xlabel(p)
    plt.ylabel(k)
    plt.show()
else:
    print(f"Columns '{p}' or '{k}' not found in the dataset.")
```
